```python
import pandas as pd
```

```python
def is_true(x: pd.Series) -> pd.Series:
    return x == "t"
```

```python
companies['iata_approved'] = is_true(companies['iata_approved'])
```

```python
companies.head()
```

```python
shuttles['d_check_complete'] = is_true(shuttles['d_check_complete'])
shuttles['moon_clearance_complete'] = is_true(shuttles['moon_clearance_complete'])
```

```python
shuttles.head()
```

```python
def  _parse_money(x: pd.Series) -> pd.Series:
    return x.astype(str).str.replace("$", "").str.replace(",", "").astype(float)
```

```python
shuttles['price'] = _parse_money(shuttles['price'])
shuttles.head()
```

```python
def _parse_percentage(x: pd.Series) -> pd.Series:
    x = x.astype(str).str.replace("%", "")
    x = x.astype(float) / 100
    return x
```

```python
companies['company_rating'] = _parse_percentage(companies['company_rating'])
companies.head()
```
